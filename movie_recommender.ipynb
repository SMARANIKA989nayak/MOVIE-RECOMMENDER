we just want the the name section for applying the tags so we need to extract the tags from the list

'{{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}}'

```python
def convert(obj):
  l = []
  for i in obj:
    l.append(i['name'])
  return l
```

```python
convert([{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}])
#this is a case of dictionary
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert1(obj1):
  l1 = []
  for i in ast.literal_eval(obj1):
    l1.append(i['name'])
  return l1
```

```python
movies['genres'] = movies['genres'].apply(convert1)
```
